Getting the data

```python
X_train, Y_train, X_test, Y_test, factor, indice, X_forecast = get_data_model(
    'hosp_GE',
    ['GE'],
    ['foph_test', 'foph_cases', 'foph_hosp'],
    ini_date = '2020-10-01', 
    look_back=21,
    predict_n=14,
    split=0.8,
    vaccine=True,
    smooth=True,
    updated_data=False)
```

```python
print('Shape of the returned values:')
print('- X_train:', X_train.shape)
print('- Y_train:', Y_train.shape)
print('- X_test:', X_test.shape)
print('- Y_test:', Y_test.shape)
print('- factor (it\'s a float value):', factor)
print('- indice:', len(indice))
print('- X_forecast:', X_forecast.shape)
```

Getting the model

```python
%%time 

model = build_model(hidden = 10, features = X_train.shape[2], predict_n = 14, look_back = 21, batch_size = 1)
```

```python
%%time
hist = train(model,
    X_train,
    Y_train,
    batch_size=1,
    epochs=100,
    path=None,
    label_history = "history_GE",
    label_model="trained_model_GE",
    save=True,)
```

Plot the training history 

```python
plot_training_history(hist, ['loss', 'val_loss'], save = True )
```

Make predictions:


```python
%%time
predicted = predict(model, np.concatenate((X_train, X_test), axis = 0), uncertainty=True)
```

```python
plot_predicted_vs_data(
    predicted,
    np.concatenate((Y_train, Y_test), axis = 0),
    indice,
    title = 'Predictions hospitalizations in GE', 
    label = 'plot_pred_teste' ,
    factor = factor,
    xlabel = 'time', 
    ylabel = 'incidence', 
    split_point= len(Y_train),
    uncertainty=True,
    save = False, 
    path = None)
```

```python
metrics = calculate_metrics(predicted, np.concatenate((Y_train, Y_test), axis = 0), factor, uncertainty = True)

metrics
```
